# sample predict

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load gambar dan convert ke RGB
img = Image.open('/content/drive/MyDrive/Colab Notebooks/tugas akhir/jahe.jpeg')
img_converted_to_rgb = img.convert('RGB')

# Resize ke ukuran kecil agar mudah ditampilkan misal 4x4 (sesuaikan sesuai kebutuhan)
img_small = img_converted_to_rgb.resize((4, 4))

# Convert ke numpy array
img_array = np.array(img_small)

# Pisahkan channel R,G,B
red_channel = img_array[:, :, 0]
green_channel = img_array[:, :, 1]
blue_channel = img_array[:, :, 2]

# Tampilkan gambar asli menggunakan matplotlib (inline display)
plt.imshow(img_converted_to_rgb)
plt.axis('off') # hilangkan axis supaya lebih bersih tampilannya
plt.show()

print("Red Channel:")
print(red_channel)

print("\nGreen Channel:")
print(green_channel)

print("\nBlue Channel:")
print(blue_channel)
```
